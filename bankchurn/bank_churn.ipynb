Sure! Here is an explanation of each column:

- **id**: Unique identifier for each record in the dataset.
- **CustomerId**: Unique identifier for each customer.
- **Surname**: Last name of the customer.
- **CreditScore**: Numerical value representing the credit score of the customer.
- **Geography**: Country or region where the customer is located.
- **Gender**: Gender of the customer.
- **Age**: Age of the customer.
- **Tenure**: Number of years the customer has been with the bank.
- **Balance**: Amount of money in the customer's account.
- **NumOfProducts**: Number of bank products the customer has.
- **HasCrCard**: Binary value indicating if the customer has a credit card (1 for yes, 0 for no).
- **IsActiveMember**: Binary value indicating if the customer is an active member (1 for yes, 0 for no).
- **EstimatedSalary**: Estimated annual salary of the customer.
- **Exited**: Binary value indicating if the customer has exited the bank (1 for yes, 0 for no).

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
rf_params = {
    'n_estimators' : [x for x in range(50,251,50)],
    'criterion' : ['gini','entropy']
}

rf = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gcv = GridSearchCV(rf,rf_params)
gcv.fit(X_train,y_train)
pred = gcv.predict(X_test)

print('accuracy_score : \n',accuracy_score(pred,y_test))
print('confusion_matrix : \n',confusion_matrix(pred,y_test))
print('classification report \n: ',classification_report(pred,y_test))
```
